```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pathlib import Path
import seaborn as sns
```

```python
gross_output_df=pd.read_excel('Resources/grossoutput.xlsx')
gross_output_df
```

```python
num_jobs_df=pd.read_csv("Resources/num_jobs.csv",encoding='latin-1',nrows=104874)
num_jobs_df
```

```python
gross_output_df = gross_output_df.rename(columns={"Unnamed: 0":"Industry"})
gross_output_df
```

```python
gross_output_df
```

```python
gross_output_df["Industry"]=gross_output_df["Industry"].str.strip()
gross_output_df
```

```python
gross_output_df.count()
```

```python
gross_output_df = gross_output_df.dropna(how='any')
```

```python
gross_output_df.count()
```

```python
num_jobs_df=num_jobs_df.loc[num_jobs_df["GeoName"]=="United States"]
num_jobs_df
```

```python
num_jobs_df=num_jobs_df.drop(columns=["GeoFIPS","Region","TableName","LineCode","IndustryClassification","Unit","2001","2002","2003",
                          "2004","2005","2006","2007","2008","2009","2010","2011","2012","2013","2014","2015","2016","2017","GeoName"])
num_jobs_df
```

```python
num_jobs_df.count()
```

```python
num_jobs_df = num_jobs_df.rename(columns={"Description":"Industry"})
num_jobs_df
```

```python
num_jobs_df.dtypes
```

```python
num_jobs_df["Industry"]=num_jobs_df["Industry"].str.strip()
num_jobs_df
```

```python
merged_df=pd.merge(gross_output_df,num_jobs_df,on="Industry")
merged_df
```

```python
merged_df.dtypes
```

```python
merged_df["2018_y"] = merged_df["2018_y"].astype(int)
merged_df["2019_y"] = merged_df["2019_y"].astype(int)
merged_df["2020_y"] = merged_df["2020_y"].astype(int)
merged_df["2021_y"] = merged_df["2021_y"].astype(int)
merged_df.dtypes
```

```python
merged_df.rename(columns={"2018_x": "Gross_output_2018",
                         "2019_x":"Gross_output_2019",
                        "2020_x":"Gross_output_2020",
                        "2021_x":"Gross_output_2021",
                         "2018_y":"Number_of_jobs_2018",
                        "2019_y":"Number_of_jobs_2019",
                        "2020_y":"Number_of_jobs_2020",
                         "2021_y":"Number_of_jobs_2021"}, inplace=True)


merged_df
```

```python
merged_df.count()
```

```python
# merged_df.drop(columns="index",inplace=True)
# merged_df
```

```python
filter_df=merged_df["Industry"].isin(["Transportation and warehousing","Arts, entertainment, and recreation",
                                     "Accommodation and food services","Health care and social assistance",
                                      "Construction","Finance and insurance","Wholesale trade","Retail trade"])
merged_df = merged_df[filter_df]
# gross_output_df
merged_df
```

```python
merged_df = merged_df.set_index("Industry")
# merged_df["Gross_output_2018"]=merged_df["Gross_output_2018"].map("${:,.2f}".format)
# merged_df["Gross_output_2018"] = merged_df["Gross_output_2018"].astype(float)
merged_df
```

```python
clean_gross_output=merged_df.filter(items=["Gross_output_2018","Gross_output_2019","Gross_output_2020","Gross_output_2021"])
clean_gross_output
```

```python
selected_columns = ['Gross_output_2018', 'Gross_output_2019', 'Gross_output_2020', 'Gross_output_2021']
df_selected = clean_gross_output[selected_columns]

# Transpose the dataframe to have years as columns
df_selected = df_selected.transpose()

# Plotting the grouped bar chart
plt.figure(figsize=(10, 6))
df_selected.plot(kind='bar', width=0.8)
plt.xlabel('Year')
plt.ylabel('Gross Output')
plt.title('Gross Output by Industry during the COVID-19')
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
locs, labels = plt.xticks()
plt.xticks(np.arange(4),['2018','2019','2020','2021'],rotation=0)
plt.grid(ls='--')
plt.tight_layout()
plt.savefig("Images/Bargraph.png")
plt.show()
```

```python
cleaned_Number_jobs=merged_df.filter(items=["Number_of_jobs_2018","Number_of_jobs_2019","Number_of_jobs_2020","Number_of_jobs_2021"])
cleaned_Number_jobs
```

```python
# Create a line chart
plt.figure(figsize=(40, 30))
cleaned_Number_jobs.transpose().plot.line(lw=3)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Number of Jobs')
plt.title('Number of Jobs availale of Sectors over Time')
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
locs, labels = plt.xticks()
plt.xticks(np.arange(4),['2018','2019','2020','2021'],rotation=0)
plt.grid(axis='x')
plt.tight_layout()
plt.savefig("Images/Linechart.png")
# Show the plot
plt.show()
```

```python
merged_df['Job_Growth_2018_2019'] = merged_df['Number_of_jobs_2019'] - merged_df['Number_of_jobs_2018']
merged_df['Job_Growth_2019_2020'] = merged_df['Number_of_jobs_2020'] - merged_df['Number_of_jobs_2019']
merged_df['Job_Growth_2020_2021'] = merged_df['Number_of_jobs_2021'] - merged_df['Number_of_jobs_2020']



# Create the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(merged_df[['Job_Growth_2018_2019', 'Job_Growth_2019_2020', 'Job_Growth_2020_2021']], cmap='YlGnBu', annot=True, fmt='.1f')
plt.title('Changes in Number of Jobs in Sectors and Years ')
plt.xlabel('Years')
plt.ylabel('Industry')
plt.tight_layout()
plt.savefig("Images/Heatmap.png")
# Display the plot
plt.show()
```

```python
merged_df['Percentage_change'] = ((merged_df['Gross_output_2020'] - merged_df['Gross_output_2019']) / merged_df['Gross_output_2019']) * 100

# Create the bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(merged_df['Gross_output_2020'], merged_df.index, s=merged_df['Number_of_jobs_2019'] / 10000, c=merged_df['Percentage_change'], cmap='RdYlGn')
plt.colorbar(label='Percentage Change in Gross Output')
plt.title('Number of Jobs vs Gross Output ')
plt.xlabel('Gross Output in 2020')
plt.ylabel('Industry')
plt.grid()
plt.tight_layout()
plt.savefig("Images/BubbleChart.png")
# Display the plot
plt.show()
```
